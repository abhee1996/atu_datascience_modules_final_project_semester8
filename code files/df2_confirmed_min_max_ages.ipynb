```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
import seaborn as sns
import matplotlib;
import io;
from google.colab import files;
import matplotlib.ticker as mtick
uploaded = files.upload()
matplotlib.rcParams["figure.figsize"] = (20,10)
df1 = pd.read_csv(io.StringIO(uploaded["COVID-19_Reported_Patient_Impact_and_Hospital_Capacity_by_State.csv"].decode('ISO-8859-1')))
#<!-----------------confirmed-------------------------->

df2_confirmed_min_max = pd.DataFrame(df1
  [[
    'previous_day_admission_adult_covid_confirmed_18-19',
    'previous_day_admission_adult_covid_confirmed_20-29',
    'previous_day_admission_adult_covid_confirmed_30-39',
    'previous_day_admission_adult_covid_confirmed_40-49',
    'previous_day_admission_adult_covid_confirmed_50-59',
    'previous_day_admission_adult_covid_confirmed_60-69',
    'previous_day_admission_adult_covid_confirmed_70-79',
    'previous_day_admission_adult_covid_confirmed_80+',

    ]].sum())

df31 = df2_confirmed_min_max[0].sum()
df2_confirmed_min_max[1] = df2_confirmed_min_max[0] / df31 * 100
corr = df2_confirmed_min_max.corr()

ax = sns.barplot(data=df2_confirmed_min_max, x=df2_confirmed_min_max.index, y=df2_confirmed_min_max.columns[1])
ax.set(xlabel='Sources of Covid Patient', ylabel='Percentage of Patients by ages',
      title='Percentage of patient Getting Information About ages of patients')
ax.set_xticklabels(
   [
           'previous_day_admission_adult_covid_confirmed_18-19',
    'previous_day_admission_adult_covid_confirmed_20-29',
    'previous_day_admission_adult_covid_confirmed_30-39',
    'previous_day_admission_adult_covid_confirmed_40-49',
    'previous_day_admission_adult_covid_confirmed_50-59',
    'previous_day_admission_adult_covid_confirmed_60-69',
    'previous_day_admission_adult_covid_confirmed_70-79',
    'previous_day_admission_adult_covid_confirmed_80+',

   ],
    rotation=45)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
#<!-----------------suspected-------------------------->


# df2_suspected_min_max = pd.DataFrame(
# df1
#   [[
#     'previous_day_admission_adult_covid_suspected_18-19',
#     'previous_day_admission_adult_covid_suspected_20-29',
#     'previous_day_admission_adult_covid_suspected_30-39',
#     'previous_day_admission_adult_covid_suspected_40-49',
#     'previous_day_admission_adult_covid_suspected_50-59',
#     'previous_day_admission_adult_covid_suspected_60_69',
#     'previous_day_admission_adult_covid_suspected_70-79',
#     'previous_day_admission_adult_covid_suspected_80',
#     ]].sum())
# df31 = df2_suspected_min_max[0].sum()
# df2_suspected_min_max[1] = df2_suspected_min_max[0] / df31 * 100
# corr = df2_suspected_min_max.corr()

# #<!------------------------------------------->
# ax_suspected_min_max = sns.barplot(data=df2_suspected_min_max, x=df2_suspected_min_max.index, y=df2_suspected_min_max.columns[1])
# ax_suspected_min_max.set(xlabel='Sources of Covid Patient', ylabel='Percentage of Patients by ages',
#       title='Percentage of patient Getting Information About ages of patients')
# ax_suspected_min_max.set_xticklabels(
#    [
#            'previous_day_admission_adult_covid_suspected_18-19',
#     'previous_day_admission_adult_covid_suspected_20-29',
#     'previous_day_admission_adult_covid_suspected_30-39',
#     'previous_day_admission_adult_covid_suspected_40-49',
#     'previous_day_admission_adult_covid_suspected_50-59',
#     'previous_day_admission_adult_covid_suspected_60_69',
#     'previous_day_admission_adult_covid_suspected_70-79',
#     'previous_day_admission_adult_covid_suspected_80',

#    ],
#     rotation=45)
# ax_suspected_min_max.yaxis.set_major_formatter(mtick.PercentFormatter())
```
